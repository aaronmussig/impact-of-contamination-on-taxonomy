```python
DF_META = GtdbMetadataR207().output().read_cached()
DF_META.head()
```

```python
DF_CSS = AggregateMaxCssLevelMerged().output().read_cached()
DF_CSS.head()
```

```python
FAIL_GIDS = set(DF_CSS.index)
```

```python
from collections import defaultdict, Counter


def generate_sp_gids():
    out = defaultdict(set)
    for gid, row in DF_META.iterrows():
        out[row['species']].add(gid)
    return out

D_SP_TO_GIDS = generate_sp_gids()
```

```python
n_all_fail = 0
for sp, gids_in_sp in D_SP_TO_GIDS.items():
    n_fail = len(gids_in_sp.intersection(FAIL_GIDS))
    if n_fail == len(gids_in_sp):
        n_all_fail +=1

print(n_all_fail)
```

```python
import pandas as pd


def generate_n_sp_with_all_failed():
    out = list()
    for sp, gids in D_SP_TO_GIDS.items():
        n_fail = len(gids.intersection(FAIL_GIDS))
        if n_fail == len(gids):
            out.append({
                'species': sp,
                'n_gids': n_fail,
                'set': 'all_fail'
            })
        elif n_fail > 0:
            out.append({
                'species': sp,
                'n_gids': n_fail,
                'set': 'subset'
            })


    return pd.DataFrame(out)

DF_SP_FAILED = generate_n_sp_with_all_failed()
DF_SP_FAILED.head()
```

```python
def generate_df_with_counts():
    out = list()
    for sp, gids in D_SP_TO_GIDS.items():
        n_fail = len(gids.intersection(FAIL_GIDS))
        out.append({
            'species': sp,
            'n_gids': len(gids),
            'n_fail': n_fail,
            'pct_fail': n_fail / len(gids) * 100
        })
    return pd.DataFrame(out)

DF_SP_COUNTS = generate_df_with_counts()
DF_SP_COUNTS.head()
```

```python
# If we order the species by the largest number of failures, and count up to 50% of all genomes in R207, what is
# the size of the species clusters we're considering
tot_counted = 0
for row in DF_SP_COUNTS.sort_values(by=['n_gids'], ascending=False).itertuples():
    tot_counted += row.n_fail
    if tot_counted >= 35723 / 2:
        print(row)
        print('done')
        break
# everything to the right of that value represents half the failed genomes
```
