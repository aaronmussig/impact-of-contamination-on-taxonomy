```python
from workflow.external.gtdb_metadata import GtdbMetadataR207
from tqdm import tqdm

from workflow.gunc_helper.aggregate_max_css_level_merged import AggregateMaxCssLevelMerged
```

```python
DF_META = GtdbMetadataR207().output().read_cached()
DF_META.head()
```

```python
DF_CSS = AggregateMaxCssLevelMerged().output().read_cached()
DF_CSS.head()
```

```python
FAIL_GIDS = set(DF_CSS.index)
```

```python
from collections import defaultdict
import pandas as pd


RANKS = ['domain', 'phylum', 'class', 'order', 'family', 'genus', 'species']


def generate_data():
    data_fail = defaultdict(lambda: defaultdict(lambda: 0))
    data_all = defaultdict(lambda: defaultdict(lambda: 0))

    for gid, row in tqdm(DF_META.iterrows(), total=len(DF_META)):
        if gid in FAIL_GIDS:
            for rank in RANKS:
                data_fail[rank][row[rank]] += 1
        for rank in RANKS:
            data_all[rank][row[rank]] += 1

    rows = list()
    for rank in RANKS:
        cur_fail = data_fail[rank]
        for taxon, cnt in cur_fail.items():
            rows.append({
                'rank': rank,
                'taxon': taxon,
                'count': cnt,
                'pct': cnt / len(DF_CSS) * 100,
                'set': 'Fail',
            })
        cur_all = data_all[rank]
        for taxon, cnt in cur_all.items():
            rows.append({
                'rank': rank,
                'taxon': taxon,
                'count': cnt,
                'pct': cnt / len(DF_META) * 100,
                'set': 'R207',
            })

    df = pd.DataFrame(rows)

    return df

DF_DATA = generate_data()
DF_DATA.head()
```

```python
DF_DATA.sort_values(by=['rank', 'set', 'count', 'taxon'], ascending=[True, True, False, True], inplace=True)

def create_data_subset(top_n=5):

    df_parts = list()

    for rank in tqdm(RANKS):
        cur_df_fail = DF_DATA[(DF_DATA['rank'] == rank) & (DF_DATA['set'] == 'Fail')]

        if len(cur_df_fail) > top_n:
            other_counts = int(sum(cur_df_fail.iloc[top_n:]['count'].values))
            df_parts.append(cur_df_fail.iloc[0:top_n])
            df_parts.append(pd.DataFrame([{
                'rank': rank,
                'taxon': 'Other',
                'count': other_counts,
                'pct': other_counts / len(DF_CSS) * 100,
                'set': 'Fail'
            }]))
        else:
            df_parts.append(cur_df_fail)

        # Add the taxa from the top n failed taxa
        df_parts_tmp = pd.concat(df_parts)
        df_parts_tmp = df_parts_tmp[(df_parts_tmp['rank'] == rank) & (df_parts_tmp['set'] == 'Fail')]

        cur_df_all = DF_DATA[(DF_DATA['rank'] == rank) & (DF_DATA['set'] == 'R207')]

        fail_taxon_set = set(df_parts_tmp['taxon'].values)

        cur_df_all_named = cur_df_all[cur_df_all['taxon'].isin(fail_taxon_set)]
        df_parts.append(cur_df_all_named)

        # Calculate the other
        df_other = cur_df_all[~cur_df_all['taxon'].isin(fail_taxon_set)]
        if len(df_other) > 0:
            other_counts = int(sum(df_other['count'].values))
            df_parts.append(pd.DataFrame([{
            'rank': rank,
            'taxon': 'Other',
            'count': other_counts,
            'pct': other_counts / len(DF_META) * 100,
            'set': 'R207'
            }]))


    df = pd.concat(df_parts, ignore_index=True)
    df.sort_values(by=['rank', 'set', 'count', 'taxon'], ascending=[True, True, False, True], inplace=True)

    return df


DF_DATA_SUBSET = create_data_subset(10)
DF_DATA_SUBSET.head()
```

```python
def generate_percentile_plot(df_set):
    if df_set == 'R207':
        n = 317542
    else:
        n = 35723

    rows = list()
    for rank in RANKS:
        DF_SP_FAIL = DF_DATA[(DF_DATA['rank'] == rank) & (DF_DATA['set'] == df_set)]
        n_observed = 0
        has_printed = False
        for i, row in enumerate(DF_SP_FAIL.itertuples()):
            i += 1
            if row.taxon == 'Other':
                continue
            n_observed += row.count
            cur_pct = n_observed / n * 100
            if cur_pct >= 50 and not has_printed:
                print(rank, i)
                has_printed = True
            rows.append({
                'rank': rank,
                'Number of taxa': i,
                '% of failed genomes': cur_pct
            })

    df = pd.DataFrame(rows)
    # g = sns.lineplot(data=df, x='Number of taxa', y='% of failed genomes', hue='rank')
    # g.set(xscale='log')
    return df

print('R207')
DF_PP_R207 = generate_percentile_plot(df_set='R207')
print('\nFail')
DF_PP_FAIL = generate_percentile_plot(df_set='Fail')

```
