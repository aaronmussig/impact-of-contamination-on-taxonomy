```python
import matplotlib.pyplot as plt
import seaborn as sns

from workflow.marker.marker_congruence_fail import MarkerCongruenceFail
```

```python

DF = MarkerCongruenceFail().output().read_cached()
DF.head()
```

```python
def plot():
    fig, ax = plt.subplots(figsize=(15, 10))

    sns.histplot([float(x) for x in DF['congruence'].values], bins=50, stat='percent')

    plt.title('Congruence (%) contigs for each failed genome marker')
    ax.set_ylabel('% of markers')
    ax.set_xlabel('Congruence of contig (%)')

    plt.show()

    return


plot()
```
