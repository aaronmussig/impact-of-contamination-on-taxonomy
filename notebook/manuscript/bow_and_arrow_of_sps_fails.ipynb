```python
from workflow.external.gtdb_metadata import GtdbMetadataR207
from tqdm import tqdm
from matplotlib.ticker import ScalarFormatter, NullFormatter

import matplotlib.pyplot as plt
from workflow.gunc_helper.aggregate_max_css_level_merged import AggregateMaxCssLevelMerged
```

```python
DF_META = GtdbMetadataR207().output().read_cached()
DF_META.head()
```

```python
DF_CSS = AggregateMaxCssLevelMerged().output().read_cached()
DF_CSS.head()
```

```python
FAIL_GIDS = set(DF_CSS.index)
```

```python
from collections import defaultdict, Counter


def generate_sp_gids():
    out = defaultdict(set)
    for gid, row in DF_META.iterrows():
        out[row['species']].add(gid)
    return out

D_SP_TO_GIDS = generate_sp_gids()
```

```python
import pandas as pd


def generate_n_sp_with_all_failed():
    out = list()
    for sp, gids in D_SP_TO_GIDS.items():
        n_fail = len(gids.intersection(FAIL_GIDS))
        if n_fail == len(gids):
            out.append({
                'species': sp,
                'n_gids': n_fail,
                'set': 'all_fail'
            })
        elif n_fail > 0:
            out.append({
                'species': sp,
                'n_gids': n_fail,
                'set': 'subset'
            })


    return pd.DataFrame(out)

DF_SP_FAILED = generate_n_sp_with_all_failed()
DF_SP_FAILED.head()
```

```python
Counter([x for x in DF_SP_FAILED[DF_SP_FAILED['set'] == 'subset']['n_gids'].values])

sum([x for x in DF_SP_FAILED[DF_SP_FAILED['set'] == 'subset']['n_gids'].values])
```

```python
def generate_df_with_counts():
    out = list()
    for sp, gids in D_SP_TO_GIDS.items():
        n_fail = len(gids.intersection(FAIL_GIDS))
        out.append({
            'species': sp,
            'n_gids': len(gids),
            'n_fail': n_fail,
            'pct_fail': n_fail / len(gids) * 100
        })
    return pd.DataFrame(out)

DF_SP_COUNTS = generate_df_with_counts()
DF_SP_COUNTS.head()
```

```python
# If we order the species by the largest number of failures, and count up to 50% of all genomes in R207, what is
# the size of the species clusters we're considering
tot_counted = 0
for row in DF_SP_COUNTS.sort_values(by=['n_gids'], ascending=False).itertuples():
    tot_counted += row.n_fail
    if tot_counted >= 35723 / 2:
        print(row)
        print('done')
        break
# everything to the right of that value represents half the failed genomes
```

```python
def generate_percentiles():
    x, y = list(), list()
    for i in range(1, 1000):
        i /= 10
        n_sp_gt_fail = len(DF_SP_COUNTS[DF_SP_COUNTS['pct_fail'] >= i])
        pct_sp_in_this_set = 100 * n_sp_gt_fail/len(D_SP_TO_GIDS)
        x.append(i)
        y.append(pct_sp_in_this_set)
    return x, y

X_percentile,Y_percentile = generate_percentiles()
```

```python
def generate_joint_plot_data():

    x_vals = list()
    y_vals = list()

    total_n_sp = len(DF_SP_COUNTS)

    for i in tqdm(range(100, -1, -1)):
        n_rows = len(DF_SP_COUNTS[DF_SP_COUNTS['pct_fail'] >= i])
        x_vals.append(i)
        y_vals.append(n_rows / total_n_sp * 100)

    return x_vals, y_vals

JOINT_DATA_X, JOINT_DATA_Y = generate_joint_plot_data()
```

```python
def generate_joint_plot_data_2():

    x_vals = list()
    y_vals = list()

    total_n_gids = sum(DF_SP_COUNTS['n_gids'].values)
    print(total_n_gids)

    for i in tqdm(range(100, -1, -1)):
        n_gids = sum(DF_SP_COUNTS[DF_SP_COUNTS['pct_fail'] >= i]['n_fail'].values)
        x_vals.append(i)
        y_vals.append(n_gids / total_n_gids * 100)

    return x_vals, y_vals

JOINT_DATA_X_CNT, JOINT_DATA_Y_CNT = generate_joint_plot_data_2()
```

```python
from matplotlib.gridspec import GridSpec

def plot_sp_counts_failed():

    fig = plt.figure(figsize=(8, 6))
    gs = GridSpec(1, 2)

    ax_scatter = fig.add_subplot(gs[0, 0])
    ax_hist_y = fig.add_subplot(gs[0, 1])
    # plt.axis('equal')


    ax_scatter.scatter(x=DF_SP_COUNTS['n_gids'], y=DF_SP_COUNTS['pct_fail'], s=5, alpha=0.4)
    # ax.set_ylim(-1000, 25000)
    # ax.set_yscale('log')

    ax_scatter.set_xscale('log')
    ax_scatter.xaxis.set_major_formatter(ScalarFormatter())
    # ax.set_xlim(0, 100)
    ax_scatter.set_xlim(0, 100000)

    # plt.gca().set_aspect('equal')

    # ax_hist_y.hist(DF_SP_COUNTS['pct_fail'], orientation='horizontal', bins=100, density=True)
    ax_hist_y.plot(JOINT_DATA_Y, JOINT_DATA_X, label='Species')
    ax_hist_y.set_xlabel('Cumulative %')
    ax_hist_y.get_yaxis().set_visible(False)

    ax_hist_y.plot(JOINT_DATA_Y_CNT, JOINT_DATA_X_CNT, label='Failed genomes')

    plt.suptitle('Percentage of genomes that failed GUNC in each species cluster')
    plt.tight_layout()
    ax_scatter.set_xlabel('Number of genomes in species cluster')
    ax_scatter.set_ylabel('Percent of genomes that failed GUNC in species cluster')

    plt.legend()

    plt.show()



    return

plot_sp_counts_failed()
```
